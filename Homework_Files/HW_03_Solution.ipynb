# Problem 1

```python
m=1 #Use a mass of '1 kg' for simplicity
c = 3e8 #speed of light
v = np.linspace(0,.99)*c #Velocity
gamma= (1-(v/c)**2)**(-1/2)
KE_relativistic = (gamma-1)*m*c**2
KE_classical = 1/2*m*v**2
```

## Add Correction below (change 0 to your expression that depends on m, v, and c)

```python
correction = 3/8*m*v**4/c**2 #Add correction here (change 0 to what you found)
```

```python
KE_correction = 1/2*m*v**2 +correction 
linewidth=4
plt.plot(v/c,KE_relativistic,label = 'Relativistic',linewidth=linewidth)
plt.plot(v/c,KE_classical,label = 'Classical',linewidth=linewidth)
plt.plot(v/c,KE_correction,label = 'Classical+Correction',linewidth=linewidth)
plt.xlabel('v/c')
plt.ylabel('Kinetic Energy (J)')
plt.legend()
plt.grid()
```

# Problem 3
$$T = 2\pi \sqrt{\frac{L}{g}} \left[1 + \frac{1}{16}\theta^2 + \frac{11}{3072}\theta^4 + \dots   \right] $$

# (a) When is the small angle approximation $T = 2 \pi \sqrt{L/g}$ good within 1%?
This will occur when 
$$   \frac{1}{16} \theta^2 \lt (Error) \times 1$$ 
or 
$$\theta = \sqrt{16(Error)} = 4\sqrt{Error}$$

```python
import math
Error = 0.01 #within 1 %
theta_radians = 4*(Error)**(1/2)
theta_degrees = theta_radians*180/math.pi
print('This is good within 1% for angles less than '+ str(theta_radians) +' radians, or '+str(theta_degrees) + ' degrees.')
```

```python
Error = 0.1 #within 1 %
theta_radians = 4*(Error)**(1/2)
theta_degrees = theta_radians*180/math.pi
print('This is good within 10% for angles less than '+ str(theta_radians) +' radians, or '+str(theta_degrees) + ' degrees. Using one term.')

Second_Order = 11/3072*theta_radians**4
print('The error introduced by the next term at this angle is ' +str(Second_Order))
```

# (c) Plot

```python
theta = np.linspace(0.001,math.pi/2)
theta_deg = theta*180/math.pi
first = theta/theta
```

## Add the third correction term below (I've added the second as an example)

```python
second = 1/16*theta**2
third = 11/3072*theta**4 # Change 0 to 11/3072 theta^4
```

```python
linewidth=3
plt.plot(theta_deg, first, label = 'No correction',linewidth=linewidth)
plt.plot(theta_deg,first+second, label = 'With Second Term',linewidth=linewidth)
plt.plot(theta_deg,first+second+third, label = 'With Second/Third Terms',linewidth=linewidth)
plt.ylabel(r'$T/(2\pi \sqrt{L/g})$)')
plt.xlabel(r'Starting Angle ($\circ$)')
plt.grid()
plt.legend()
plt.show()
```
