## (a) $$y = v_0 t - gt^2/2$$

```python
t = np.linspace(0,10) #Set time from 0 to 10 units (s)

y = t  #Change this line to the right equation

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.show()
```

## (b) $$x = x_0 e^{-\frac{\alpha}{10} t} \sin(\omega t)$$

```python
t = np.linspace(0,10*np.pi,num=1000) #Set time from 0 to 20 pi units 

y = np.exp(-.1*t) #Change this line to the right equation (need to include sine term)

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('x')
plt.grid()
plt.show()
```

## (c)  $$f(x) = \frac{x}{a^3+x^3}$$

```python
x = np.linspace(-15,15,num=100) #Set time from 0 to 20 pi units 

f = x**3 #Change this line to the right equation 

plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

## (d) $$f(x) = \frac{1}{\sin(x)}, ~~ \rm{for}~ (-3\pi < x < +4\pi) $$

```python
t = np.linspace(-3*np.pi,4*np.pi,num=200) #Set time from 0 to 20 pi units 

y = 1/np.sin(t) #Ok, You don't need to change anything for this one

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('x')
plt.ylim(-10,10)
plt.grid()
plt.show()
```

## (e) $$f(x) = \tanh (x) $$

```python
x = np.linspace(-15,15,num=100) #Set time from 0 to 20 pi units 

f = np.tan(x) #Change this line to the right equation 

plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```
