# Example Integral for P 5.4 Nearing

```python
import sympy as sp
x,L = sp.symbols('x L',real=True)
m = sp.symbols('m',integer=True,positive=True)
```

```python
sp.integrate(x**2*sp.cos(m*sp.pi*x/L), (x,0,L))
```

# Graph Fourier Series Approximation

```python
num_terms = 3 #Increase to increase number of terms in the sum
xsquared = L**2/3 + 4 * L**2/sp.pi**2 * sp.Sum((-1)**m/m**2 * sp.cos(m*sp.pi*x/L),(m,1,num_terms))
p1 =sp.plot((xsquared.subs({L:1})), (x, -2, 2),show=False,label='Fourier Series')
p1[0].label = 'Fourier Series: '+ str(num_terms+1) + ' term(s)'
p2=sp.plot(x**2,(x,0,1),show=False)
p2[0].label = r'$x^2$'
p1.append(p2[0])
p1.legend=True
p1.show()
```
