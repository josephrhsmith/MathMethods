```python
#Load libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Hyperbolic Cosine (cosh)

```python
x = np.linspace(-4,4)     #Create an array of points from -4 to 4 (Default is 50 points)
y = np.cosh(x)            #Evaluate the function at all points in the domain
plt.plot(x,y,linewidth=3)

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## Plot Other Hyperbolic Trig Funcitons

```python
x = np.linspace(-4,4) 

y1 = np.cosh(x)  
plt.plot(x,y1,linewidth=3,label='cosh')

y2 = np.sinh(x)  
plt.plot(x,y2,linewidth=3,label='sinh')

#Add tanh here

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## $f(x) = \frac{x}{a^2-x^2}$

```python
x = np.linspace(-4,4)

a=1
y = x/(a**2-x**2)  
plt.plot(x,y,linewidth=3)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```
