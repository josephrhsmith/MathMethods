# Example Fourier series for $x^2$

```python
range=1
x = np.linspace(0,range,200)
plt.xlim(0,range)
L = 1
y1= L**2/3*np.ones(len(x))
y2 = -4*L**2/np.pi**2 * np.cos(np.pi*x/L)
y3 = 4*L**2/np.pi**2/4 * np.cos(2*np.pi*x/L)
y4 =  -4*L**2/np.pi**2/9 * np.cos(3*np.pi*x/L)
plt.grid()
plt.xlabel('x',fontsize='x-large')
plt.ylabel('y',fontsize='x-large')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(x,y1,linewidth=4,label =r'$L^2/3$' )
plt.plot(x,y2,linewidth=4,label =r'$-4L^2/\pi^2\cos(\pi x/L)$' )
plt.plot(x,y3,linewidth=4,label =r'$L^2/\pi^2\cos(2\pi x/L)$' )
plt.legend(fontsize='x-large',loc='lower right')
plt.title('L = ' + str(L))
plt.show()
```

```python
plt.plot(x,x**2,linewidth=4,label =r'$x^2$')
plt.grid()
plt.xlabel('x',fontsize='x-large')
plt.ylabel('y',fontsize='x-large')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(x,y1+y2,linewidth=4,label ='two terms',ls='--')
#plt.plot(x,y1+y2+y3,linewidth=4,label ='two terms',ls='--')
plt.legend()
plt.show()
```

```python
plt.plot(x,x**2,linewidth=4,label =r'$x^2$')
plt.grid()
plt.xlabel('x',fontsize='x-large')
plt.ylabel('y',fontsize='x-large')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
#plt.plot(x,y1+y2,linewidth=4,label ='two terms',ls='--')
plt.plot(x,y1+y2+y3,linewidth=4,label ='three terms',ls='--')
plt.legend()
plt.show()
```

```python
plt.plot(x,x**2,linewidth=4,label =r'$x^2$')
plt.grid()
plt.xlabel('x',fontsize='x-large')
plt.ylabel('y',fontsize='x-large')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
#plt.plot(x,y1+y2,linewidth=4,label ='two terms',ls='--')
#plt.plot(x,y1+y2+y3,linewidth=4,label ='three terms',ls='--')
plt.plot(x,y1+y2+y3+y4,linewidth=4,label ='three terms',ls='--')
plt.legend()
plt.show()
```

# Orthogonal Functions (SymPy)

```python
import sympy as sp
```

```python
x,L = sp.symbols('x L',real=True)
m,n = sp.symbols('m n',integer=True,positive=True)
```

```python
sp.integrate(sp.sin(n*sp.pi*x/L) *sp.sin(m*sp.pi*x/L), (x,0,L))
```

```python
sp.integrate(sp.cos(n*sp.pi*x/L) *sp.cos(m*sp.pi*x/L), (x,0,L))
```
